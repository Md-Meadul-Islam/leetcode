```python
from collections import Counter
```

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t): return False
        sortedS = "".join(sorted(s))
        sortedT = "".join(sorted(t))
        if sortedS == sortedT : return True
        return False
    def isAnagramWithCounter(self, s:str, t:str)-> bool:
        if len(s) != len(t): return False
        count_s = Counter(s)
        count_t = Counter(t)
        for char in count_s:
            if char not in count_t:
                return False
            elif count_t[char] != count_s[char]:
                return False
        return True
    def isAnagramWithCount(self, s:str, t:str)-> bool:
        if len(s) != len(t): return False
        for i in set(s):
            if t.count(i) != s.count(i):
                return False
        return True
```

```python
s = Solution()
s.isAnagram("anagram", "nagarm")
```

```python
s = Solution()
s.isAnagramWithCounter("anagram", "nagaarm")
```

```python
s = Solution()
s.isAnagramWithCount("anagram", "nagaarm")
```
