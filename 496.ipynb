```python
from typing import List
```

```python
#503. Next Greater Element I
class Solution:
    def nextGreaterElement2(self, nums1: List[int], nums2: List[int]) -> List[int]:
        nums:List[int] = [0] * len(nums1)
        j = len(nums2) - 1
        stack = []
        for i in range(j, -1, -1):
            while stack and stack[-1]<=nums2[i]:
                stack.pop()
            if(nums2[i] in nums1):
                idx = nums1.index(nums2[i])
                nums[idx] = stack[-1] if stack else -1
            stack.append(nums2[i])
        return nums
    def nextGreaterElement(self, nums1: List[int], nums2: List[int]) -> List[int]:
        hashmap = {}
        stack = []
        for num in nums2:
            while stack and stack[-1] <num:
                hashmap[stack.pop()] = num
            stack.append(num)

        return [hashmap.get(num, -1) for num in nums1]


```

```python
s = Solution()
s.nextGreaterElement([4,1,2], [1,3,4,2, 5])
```
