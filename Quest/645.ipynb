645. Set Mismatch

```python
from typing import List
class Solution:
    def findErrorNums(self, nums: List[int]) -> List[int]:
        hash_map = {}
        res = []
        for i in range(1, len(nums)+1):
            hash_map[i] = 0
        for num in nums:
            if num in hash_map:
                hash_map[num] +=1
            if hash_map[num] >1:
                res.append(num)
        res.append(min(hash_map, key=hash_map.get))
        return res
```

```python
obj = Solution()
obj.findErrorNums([3,2,2,4])
```
