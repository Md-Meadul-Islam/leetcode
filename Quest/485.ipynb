```python
#Q3. Max Consecutive Ones
from typing import List
class Solution:
    def findMaxConsecutiveOnes(self, nums: List[int]) -> int:
        consecutive_len = 0
        max_len = 0
        for num in nums:
            if num == 1:
                consecutive_len += 1
                max_len  = max(max_len, consecutive_len)
            else:
                consecutive_len = 0
        return max_len
```
