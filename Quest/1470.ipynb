```python
#1470. Shuffle the Array
from typing import List

class Solution:
    def shuffle(self, nums: List[int], n: int) -> List[int]:
        new_arr = []
        for i in range(n):
            new_arr.append(nums[i])
            new_arr.append(nums[i+n])
        return new_arr
    def shuffle2(self, nums: List[int], n: int) -> List[int]:
        new_arr = []
        arr_1 = nums[:n]
        arr_2 = nums[n:]
        for i in range(n):
            new_arr.append(arr_1[i])
            new_arr.append(arr_2[i])
        return new_arr
```

```python
obj = Solution()
obj.shuffle([2,5,1,3,4,7], 3)
```
