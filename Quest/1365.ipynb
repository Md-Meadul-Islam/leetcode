1365. How Many Numbers Are Smaller Than the Current Number

```python
from typing import List
class Solution:
    def smallerNumbersThanCurrent(self, nums: List[int]) -> List[int]:
        res = [0] * len(nums)
        hash_map = {}

        # copy original nums because we will sort
        original = nums[:]

        nums.sort()

           # store first index of each number
        for i in range(len(nums)):
            if nums[i] not in hash_map:
                hash_map[nums[i]] = i

        # build result using original order
        print(hash_map)
        for i in range(len(original)):
            res[i] = hash_map[original[i]]

        return res
    def smallerNumbersThanCurrent2(self, nums):
        n= len(nums)
        res = [0]*n
        hash_map = {}
        original = nums[:]
        nums.sort()
        for i in range(n):
            if nums[i] not in hash_map:
                hash_map[nums[i]] = i
        for i in range(n):
            res[i] = hash_map[original[i]]
        print(hash_map)
        return res

        
```
