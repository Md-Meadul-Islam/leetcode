448. Find All Numbers Disappeared in an Array

```python
from typing import List
class Solution:
    def findDisappearedNumbers(self, nums: List[int]) -> List[int]:
        hash_map = {}
        res = []
        for i in range(1, len(nums)+1):
            hash_map[i] = 0
        for num in nums:
            if num in hash_map:
                hash_map[num] +=1
        for key in hash_map:
            if hash_map[key] == 0:
                res.append(key)
        return res
    
    def findDisappearedNumbers2(self, nums: List[int]) -> List[int]:
        res = []

        for i in range(len(nums)):
            idx = abs(nums[i]) - 1
            if nums[idx] > 0:
                nums[idx] = -nums[idx]

        for i in range(len(nums)):
            if nums[i] > 0:
                res.append(i + 1)

        return res
    
    def findDisappearedNumbers3(self, nums: List[int]) -> List[int]:
        return list(set(range(1,len(nums)+1)) - set(nums))
    
    def findDisappearedNumbers4(self, nums: List[int]) -> List[int]:
        d = set(nums)

        return [cc for cc in range(1,len(nums)+1) if cc not in d]
```

```python
obj = Solution()
obj.findDisappearedNumbers2([4,3,2,7,8,2,3,1])
```
