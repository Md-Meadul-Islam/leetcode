337. House Robber III

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

from typing import Optional

class Solution:
    def rob(self, root: Optional[TreeNode]) -> int:
        if not root:
            return 0
        dp = {}
        def dfs(node):
            if not node:
                return 0
            if node in dp:
                return dp[node]
            pick = node.val
            not_pick = 0
            if node.left:
                pick += dfs(node.left.left) + dfs(node.left.right)
                not_pick += dfs(node.left)
            if node.right:
                pick += dfs(node.right.right) + dfs(node.right.left)
                not_pick += dfs(node.right)
            dp[node] = max(pick, not_pick)         
            return dp[node]
        return dfs(root)
        
```

```python
tree = TreeNode(3)
tree.left = TreeNode(2)
tree.right = TreeNode(3)
tree.left.right = TreeNode(3)
tree.right.right = TreeNode(1)
```

```python
obj = Solution()
obj.rob(tree)
```
