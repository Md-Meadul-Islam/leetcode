```python
from typing import List
```

```python
class Solution:
    def subarraySum(self, nums: List[int], k: int) -> int:
        count = 0
        hash_map = {0:1}
        curr_sum = 0
        for i in  range(len(nums)):
            curr_sum += nums[i]
            prefix = curr_sum - k
            if prefix in hash_map:
                count += hash_map[prefix]
            hash_map[curr_sum] = hash_map.get(curr_sum, 0) + 1
        return count

        
```

```python
obj = Solution()
print(obj.subarraySum([1], 0))
```
