```python
from typing import List
```

```python
class Solution:
    def rob(self, nums: List[int]) -> int:
        def recur(idx,end)->int:
            if( idx>end): return 0
            pick =nums[idx] + recur(idx+2,end)
            not_pick = recur(idx+1,end)
            return max(pick, not_pick)
        n = len(nums)
        if n==0:return 0
        if n==1:return nums[0]
        ans1 = recur(0, len(nums)-2)
        ans2 = recur(1, len(nums)-1)
        return max(ans1, ans2)
    
    def robMemoization(self, nums:List[int])->int:
        def recur(idx:int,end:int, dp:dict[int, int])->int:
            if( idx>end): return 0
            if(idx in dp): return dp[idx]
            pick =nums[idx] + recur(idx+2,end,dp)
            not_pick = recur(idx+1,end, dp)
            dp[idx] = max(pick, not_pick)
            return dp[idx]
        n = len(nums)
        if n==0:return 0
        if n==1:return nums[0]
        dp1 = {}
        dp2 = {}
        ans1 = recur(0, len(nums)-2, dp1)
        ans2 = recur(1, len(nums)-1, dp2)
        return max(ans1, ans2)
    
    def robTabulation(self, nums:List[int])->int:
        n= len(nums)
        if n==0:return 0
        if n==1:return nums[0]
         
        def tabulation(start:int, end:int)->int:
            prev2 = 0 #n-2
            prev1 = 0 #n-1
            for i in range(start, end+1):
                curr = max(prev1, nums[i]+ prev2)
                prev2 = prev1
                prev1 = curr
            return prev1
        pick = tabulation(0, n-2)
        not_pick = tabulation(1, n-1)
        return max(pick, not_pick)

```

```python
obj = Solution()
nums = [1,2,3]
obj.robTabulation(nums)
```
