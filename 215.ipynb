215. Kth Largest Element in an Array

```python
from typing import List
import heapq, random

class Solution:
    def findKthLargest_heap(self, nums: List[int], k: int) -> int:
        heap = []
        n= len(nums)
        for i in range(n):
            heapq.heappush(heap, nums[i])
            print(heap)
            if len(heap)>k:
                heapq.heappop(heap)
        return heap[0]
    
    def findKthLargest_offset(self, nums: List[int], k: int) -> int:
        offset = 10000
        freq = [0]*20001
        for x in nums:
            freq[x+offset] += 1
        for i in range(20000, -1, -1):
            k -= freq[i]
            if k <= 0:
                return i-offset
    
    def findKthLargest_quick_select(self, nums: List[int], k: int) -> int:
        n = len(nums)
        target = n - k   # kth largest â†’ index in sorted order
        
        def quickselect(left, right):
            # Random pivot avoids worst case
            pivot_index = random.randint(left, right)
            pivot_value = nums[pivot_index]
            
            # Move pivot to end
            nums[pivot_index], nums[right] = nums[right], nums[pivot_index]
            
            store_index = left
            
            # Partition
            for i in range(left, right):
                if nums[i] < pivot_value:
                    nums[i], nums[store_index] = nums[store_index], nums[i]
                    store_index += 1
            
            # Move pivot to final place
            nums[store_index], nums[right] = nums[right], nums[store_index]
            
            # Divide & Conquer decision
            if store_index == target:
                return nums[store_index]
            elif store_index < target:
                return quickselect(store_index + 1, right)
            else:
                return quickselect(left, store_index - 1)
        
        return quickselect(0, n - 1)
    
    def findKthLargest(self, nums, k):
        target = len(nums) - k
        left, right = 0, len(nums) - 1
        
        while True:
            pivot_index = random.randint(left, right)
            pivot = nums[pivot_index]
            
            nums[pivot_index], nums[right] = nums[right], nums[pivot_index]
            
            store = left
            for i in range(left, right):
                if nums[i] < pivot:
                    nums[i], nums[store] = nums[store], nums[i]
                    store += 1
            
            nums[store], nums[right] = nums[right], nums[store]
            
            if store == target:
                return nums[store]
            elif store < target:
                left = store + 1
            else:
                right = store - 1

```

```python
obj = Solution()
nums = [3,2,3,1,2,4,5,5,6]#[1,2,2,3,3,4,5,5,6]
# nums = [3,2,1,5,6,4]
k = 4
obj.findKthLargest(nums, k)
```
