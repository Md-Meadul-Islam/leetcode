215. Kth Largest Element in an Array

```python
from typing import List
import heapq
class Solution:
    def findKthLargest(self, nums: List[int], k: int) -> int:
        heap = []
        n= len(nums)
        for i in range(n):
            heapq.heappush(heap, nums[i])
            print(heap)
            if len(heap)>k:
                heapq.heappop(heap)
        return heap[0]
        

```

```python
obj = Solution()
nums = [3,2,3,1,2,4,5,5,6]#[1,2,2,3,3,4,5,5,6]
# nums = [3,2,1,5,6,4]
k = 4
obj.findKthLargest(nums, k)
```
