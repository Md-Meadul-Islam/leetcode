```python
from typing import List
class Solution:
    def subsetsWithDup(self, nums: List[int]) -> List[List[int]]:
        nums.sort()
        n = len(nums)
        fin_arr = []
        def backtrack(i, res):#0,1,2
            fin_arr.append(res[:])
             
            while i < n:
                res.append(nums[i])
                backtrack(i+1, res)
                res.pop()

                val = nums[i]
                while i < n and nums[i] == val:
                    i += 1

        backtrack(0, [])
        return fin_arr
    def subsetsWithDup_2(self, nums:List[int])->List[List[int]]:
        nums.sort()
        n = len(nums)
        fin_arr = []
        def backtrack(i:int, res:List[int]):
            if i == n:
                fin_arr.append(res[:])
                return
            #include
            res.append(nums[i])
            backtrack(i+1, res)
            #exclude
            res.pop()
            idx = i+1
            while idx < n and nums[idx-1] == nums[idx]:#skipping while curr nums match with next one
                idx += 1
            backtrack(idx, res)
        backtrack(0, [])
        return fin_arr
```
