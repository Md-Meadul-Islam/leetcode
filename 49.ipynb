```python
class Solution:
    def groupAnagrams(self, strs:[]):
        hashmap = {}
        for str in strs:
            print(sorted(str))
            sortedStr = "".join(sorted(str))
            if sortedStr not in hashmap:
                hashmap[sortedStr] = []
            hashmap[sortedStr].append(str)
        return list(hashmap.values())

```

```python
something =  Solution()
```

```python
something.groupAnagrams(['abd', "dab"])
```
