```python
from typing import List
import bisect 

class Solution:
    def minSubArrayLen(self, target: int, nums: List[int]) -> int:
        left =0
        right = 0
        curr_sum = 0
        min_length = float('inf')
        while right < len(nums):
            curr_sum += nums[right]
            while curr_sum >= target:
                min_length = min(min_length, right-left+1)
                curr_sum -= nums[left]
                left += 1
            right += 1
        return min_length if min_length != float('inf') else 0
    
    def minSubArrayLen_PrefixSum(self, target:int, nums:List[int])->int:
        n = len(nums)
        prefix_sum = [0]*(n+1)
        min_length = float('inf')

        for i in range(n):
            prefix_sum[i+1] = prefix_sum[i] + nums[i]
        for i in range(n):
            need = target + prefix_sum[i]
            j = bisect.bisect_left(prefix_sum, need)
            if j<=n:
                min_length = min(min_length, j-i)
        return min_length if min_length != float('inf') else 0
```
