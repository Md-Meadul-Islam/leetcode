```python
class Solution:
    def minCut(self, s: str) -> int:
        def isPalindrome(l:int, r:int):
            while l<r:
                if s[l] != s[r]:
                    return False
                l += 1
                r -= 1
            return True
        n = len(s)
        def backtrack(start):
            if start == n:
                return 0
            min_cut = n-1
            
            for end in range(start, n):
                if isPalindrome(start, end):
                    cost = 1 + backtrack(end+1)
                    min_cut = min(min_cut, cost)
            return min_cut
        return backtrack(0)
    
        
```
