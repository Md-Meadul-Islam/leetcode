```python
from typing import List
```

```python
class Solution:
    def nextPermutation(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        last = nums[-1]
        secondLast = nums[-2]
        nums[-2] = last
        nums[-1] = secondLast
        print(nums)
```

```python
s = Solution()
s.nextPermutation([1,2,3])
```
