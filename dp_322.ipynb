322. Coin Change

```python
def coinChange(coin, amount:int):
    def dfs(rem:int):
        if rem == 0:
            return 0
        if rem <0:
            return float('inf')
        curr_count = float('inf')
        rest_amount = dfs(rem - coin)
        if rest_amount != float('inf'):
            curr_count = min(curr_count, rest_amount+1)
        return curr_count
    result = dfs(amount)
    return -1 if result == float('inf') else result
coinChange(2, 10)
```

```python
from typing import List

class Solution:
    def coinChange(self, coins:List[int], amount:int):
        def dfs(rem):
            if rem == 0:
                return 0
            if rem < 0:
                return float('inf')

            ans = float('inf')
            for c in coins:
                res = dfs(rem - c)
                if res != float('inf'):
                    ans = min(ans, res + 1)

            return ans

        result = dfs(amount)
        return -1 if result == float('inf') else result      
    
    def coinChange_memoization(self, coins:List[int], amount:int):
        memo = {}
        def dfs(rem):
            if rem == 0:
                return 0
            if rem < 0:
                return float('inf')
            if rem in memo:
                print(rem)
                return memo[rem]
            ans = float('inf')
            for c in coins:
                res = dfs(rem - c)
                if res != float('inf'):
                    ans = min(ans, res + 1)
            memo[rem] = ans
            return ans
        print(memo)
        result = dfs(amount)
        return -1 if result == float('inf') else result     
     
    def coinChange_tabulation(self, coins, amount):
        INF = float('inf')
        dp = [INF] * (amount + 1)
        dp[0] = 0

        for i in range(1, amount + 1):
            for c in coins:
                if i - c >= 0:
                    dp[i] = min(dp[i], dp[i - c] + 1)
        print(dp)
        return -1 if dp[amount] == INF else dp[amount]
    
```
