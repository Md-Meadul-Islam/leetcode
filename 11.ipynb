```python
from typing import List
```

```python
class Solution:
    def maxArea(self, height: List[int]) -> int:
        left =0
        right = len(height) - 1
        maxArea = 0
        while left<right:
            minHeightBar = min(height[left], height[right])
            distance = right - left
            area = minHeightBar * distance
            maxArea = max(maxArea, area)
            if height[left]<height[right]:
                left += 1
            else:
                right -= 1
        return maxArea
```

```python
s = Solution()
s.maxArea([23, 34, 54, 65, 67,1])
```
