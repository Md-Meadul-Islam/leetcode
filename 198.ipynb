```python
from typing import List
```

```python
class Solution:
    def rob(self, nums: List[int]) -> int:
        n = len(nums)
        if n == 0: return 0
        if n == 1: return nums[0]

        def recur(i:int, end:int)->int:
            if i > end: return 0
            pick = nums[i] + recur(i+2, end)
            not_pick = recur(i+1, end)
            return max(pick, not_pick)
        return recur(0, n-1)
    def robMemoization(self, nums:List[int])->int:
        n = len(nums)
        if n == 0 : return 0
        if n == 1 : return nums[0]

        dp = {}
        def recur(i:int, end:int)->int:
            if i > end: return 0
            if i in dp: return dp[i]
            pick = nums[i] + recur(i+2, end)
            not_pick = recur(i+1, end)
            dp[i]= max(pick, not_pick)
            return dp[i]
        return recur(0, n-1)
    def robTabulation(self, nums:List[int])->int:
        n = len(nums)
        if n == 0: return 0
        if n == 1: return nums[0]
        prev2 = 0#n-2
        prev1 = 0#n-1
        for i in range(0, n):
            curr = max(prev1, nums[i]+prev2)
            prev2 = prev1
            prev1 = curr
        return prev1
        
```

```python
obj = Solution()
obj.robTabulation([2,7,9,3,1])
```
