```python
from typing import List
class Solution:
    def subarraySum(self, nums: List[int], k: int) -> int:
        sub_arrays = 0
        dp = [0]*len(nums)
        dp[0] = nums[0]
        for i in range(1, len(nums)):
            dp[i] = nums[i]
            if(dp[i]==k):
                sub_arrays += 1
            curr_sum = dp[i-1]+dp[i]
            if curr_sum == k:
                sub_arrays += 1
            elif curr_sum <k:
                dp[i] += curr_sum
        print(dp)
        return sub_arrays

        
```

```python
obj = Solution()
obj.subarraySum([3, 4, 7, -2, 2, 1, 4, 2], 7)
```
