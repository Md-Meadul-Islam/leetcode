```python
from typing import List
class Solution:
    def __init__(self):
        self.final_pallindrome:List[List[str]] = []
    def isPalindrome(self, l:int, r:int, s:str):
        while l<r:
            if s[l] != s[r]:
                return False
            l += 1
            r -= 1
        return True
    def backtrack(self, s:str, start:int, path:List[str]):
        n = len(s)
        if start == n:
            self.final_pallindrome.append(path[:])
            return
        for end in range(start, n):
            curr_str = s[start:end+1]
            if self.isPalindrome(start, end, s):
                path.append(curr_str)
                self.backtrack(s, end+1, path)
                path.pop()

    def partition(self, s: str) -> List[List[str]]:
        self.backtrack(s, 0, [])
        return self.final_pallindrome
        
```
