```python
from typing import List


class Solution:
    def rob(self, nums: List[int]) -> int:
        n = len(nums)
        if n == 0:
            return 0
        if n == 1:
            return nums[0]
        def dfs(i, end):
            if i >end:
                return 0
            pick = nums[i]+dfs(i+2, end)
            not_pick = dfs(i+1, end)
            return max(pick, not_pick)
        rob1 = dfs(0, n-2)
        rob2 = dfs(1, n-1)
        return max(rob1, rob2)
    def rob2(self, nums:List[int])->int:
        n = len(nums)
        if n == 0:
            return 0
        if n == 1:
            return nums[0]
        dp1 = {}
        dp2 = {}

        def dfs(i, end, dp):
            if i>end:
                return 0
            pick = nums[i]+dfs(i+2, end, dp)
            not_pick = dfs(i+1, end, dp)
            dp[i] = max(pick, not_pick)
            return dp[i]
        
        ans1 = dfs(0, n-2, dp1)
        ans2 = dfs(1, n-1, dp2)
        return max(ans1, ans2)
```
