```python
from typing import List
```

```python
class Solution:
    def maxArea(self, height: List[int]) -> int:
        left =0
        right = len(height)-1
        max_area = 0
        while left<right:
            min_height = min(height[left], height[right])
            distance = right - left
            max_area = max(max_area, min_height*distance)
            if height[left]>=height[right]:
                right -= 1
            else:
                left += 1
        return max_area
            


```

```python
obj = Solution()
obj.maxArea([1,8,6,2,5,4,8,3,7])
```
