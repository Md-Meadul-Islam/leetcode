```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        hash_map = {}
        left = 0
        counter = 0
        for i in range(len(s)):
            ch = s[i]
            if ch in hash_map and hash_map[ch] >=left:
                left = hash_map[ch] + 1
            counter = max(counter, i-left+1)
            hash_map[ch] = i
        return counter
```

```python
obj = Solution()
obj.lengthOfLongestSubstring("abcabcbb")
```
