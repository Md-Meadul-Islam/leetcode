```python
from typing import List
import bisect
class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:
        n = len(nums)
        if n == 0:
            return 0
        if n == 1:
            return 1
        dp = [1]*n
        max_len = 1
        for i in range(1,n):
            for j in range(i):
                if nums[i]>nums[j]:
                    dp[i]=max(dp[i], dp[j]+1)
                    max_len = max(dp[i], max_len)
        return max_len
    def lengthOfLIS2(self, nums:List[int])->int:
        tails:List[int] = []
        for num in nums:
            idx = bisect.bisect_left(tails, num)
            print(idx)
            if idx == len(tails):
                tails.append(num)
            else:
                tails[idx] = num
```
