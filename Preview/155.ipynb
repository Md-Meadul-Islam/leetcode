```python
class MinStack:
    def __init__(self)->None:
        self.stack= []

    def push(self, val:int)->None:
        if not self.stack:
            self.stack.append((val, val))
        else:
            prev_min = self.stack[-1][1]
            curr_min = min(prev_min, val)
            self.stack.append((val, curr_min))
    def pop(self):
        if self.stack:
            self.stack.pop()
    def top(self):
        if self.stack:
            top_elm = self.stack[-1]
            return top_elm[0]
    def getMin(self):
        if self.stack:
            return self.stack[-1][1]
```
