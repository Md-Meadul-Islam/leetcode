1877. Minimize Maximum Pair Sum in Array

```python
from typing import List

class Solution:
    def minPairSum(self, nums: List[int]) -> int:
        nums.sort()
        n = len(nums)
        left = 0
        right = n-1
        max_sum = 0
        while left<right:
            curr_sum = nums[left]+nums[right]
            max_sum = max(max_sum, curr_sum)
            left += 1
            right -= 1
        return max_sum
```
