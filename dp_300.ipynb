300. Longest Increasing Subsequence

```python
from typing import List
import bisect
```

```python
class Solution:
    def lengthOfLIS(self, nums: List[int])->int :
        n = len(nums)
        if n == 0: return 0
        if n == 1: return 1
        dp = [1]*n#[1, 1, 1, 2, 2, 3, 3, 4, 3]
                       #[0,1, 2, 3, 4, 5, 6, 7, 8]
        max_len = 1
        for i in range(1, n):#7, 18
            for j in range(i):#4,3
                if nums[i]>nums[j]:
                    dp[i] =max(dp[i], dp[j]+1)
                    max_len = max(max_len, dp[i]) 
        return max_len        
    
    def lengthOfLIS_BinarySearch(self, nums: List[int])->int:
        sub = []
        for num in nums:
            idx = bisect.bisect_left(sub, num)
            if idx == len(sub):#addition
                sub.append(num)
            else:
                sub[idx] = num#replace
        return len(sub)

        
```
