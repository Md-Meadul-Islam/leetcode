```python
class MinStack:

    def __init__(self):
        self.stack = []
        

    def push(self, val: int) -> None:
        if not self.stack:
            self.stack.append((val, val))
        else:
            minVal = min(val, self.stack[-1][1])
            self.stack.append((val, minVal))
        

    def pop(self) -> None:
        self.stack.pop()

    def top(self) -> int:
        return self.stack[-1][0]
        

    def getMin(self) -> int:
        return self.stack[-1][1]
```

```python
obj = MinStack()
```

```python
obj.push(347384)
```
