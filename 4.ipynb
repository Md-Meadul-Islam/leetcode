4. Median of Two Sorted Arrays

```python
from typing import List
class Solution:
    def findMedianSortedArrays_brute_force(self, nums1: List[int], nums2: List[int]) -> float:
        nums3 = []
        i, j = 0, 0
        n1= len(nums1)
        n2 = len(nums2)
        while i<n1 and j<n2:
            if nums1[i] < nums2[j]:
                nums3.append(nums1[i])
                i += 1
            else:
                nums3.append(nums2[j])
                j += 1
        while i < n1:
            nums3.append(nums1[i])
            i += 1
        while j < n2:
            nums3.append(nums2[j])
            j += 1
        total_len = n1 + n2
        idx = int(total_len/2)
        if total_len % 2 == 1:
            return float(nums3[idx])
        else:
          return  nums3[idx-1] + nums3[idx] / 2
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        n1 = len(nums1)
        n2 = len(nums2)
        n = n1 + n2
        mid = int(n/2)

        search_idx1 = mid - 1
        search_idx2 = mid
        search_num1 = -1
        search_num2 = -1

        i, j, counter = 0, 0, 0

        while i < n1 and j < n2 and counter <= search_idx2:
            if nums1[i]<nums2[j]:
                if counter == search_idx1:
                    search_num1 = nums1[i]
                if counter == search_idx2:
                    search_num2 = nums1[i]
                i += 1
            else:
                if counter == search_idx2:
                    search_num2 = nums2[j]
                if counter == search_idx1:
                    search_num1 = nums2[j]
                j += 1
            counter += 1
        while i <n1 and counter <= search_idx2:
            if counter == search_idx1:
                search_num1 = nums1[i]
            if counter == search_idx2:
                search_num2 = nums1[i]
            counter += 1
            i += 1
        while j<n2 and counter <= search_idx2:
            if counter == search_idx2:
                search_num2 = nums2[j]
            if counter == search_idx1:
                search_num1 = nums2[j]
            counter += 1
            j += 1
        if n % 2 == 1:
            return search_num2
        else:
            return (search_num1 + search_num2) /2

    def findMedianSortedArrays_binary_search(self, nums1: List[int], nums2: List[int]) -> float:
        if len(nums1) > len(nums2):
            nums1, nums2 = nums2, nums1
        
        n1, n2 = len(nums1), len(nums2)
        low, high = 0, n1
        
        while low <= high:
            cut1 = (low + high) // 2
            cut2 = (n1 + n2 + 1)//2 - cut1
            
            left1 = nums1[cut1-1] if cut1 > 0 else float('-inf')
            left2 = nums2[cut2-1] if cut2 > 0 else float('-inf')
            
            right1 = nums1[cut1] if cut1 < n1 else float('inf')
            right2 = nums2[cut2] if cut2 < n2 else float('inf')
            
            if left1 <= right2 and left2 <= right1:
                if (n1 + n2) % 2 == 0:
                    return (max(left1, left2) + min(right1, right2)) / 2
                else:
                    return max(left1, left2)
            elif left1 > right2:
                high = cut1 - 1
            else:
                low = cut1 + 1



        
```
