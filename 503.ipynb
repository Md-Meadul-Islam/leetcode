```python
#503. Next Greater Element II
from typing import List
```

```python
# make the array double to find the elm circularly
class Solution:
    def nextGreaterElements2(self, nums: List[int]) -> List[int]:
        n = len(nums)
        stack = []
        res = [-1]* n
        for i in range(2*n-1, -1, -1):
            num = nums[i%n]
            while stack and nums[stack[-1]] <= num:
                stack.pop()
            res[i%n] = nums[stack[-1]] if stack else -1
            stack.append(i%n)
        return res
    def nextGreaterElements(self, nums: List[int]) -> List[int]:
        n = len(nums)
        stack = []
        res = [-1]* n
        for i in range(2*n-1):
            num = nums[i%n]
            while stack and nums[stack[-1]] < num:
                res[stack.pop()] = num

            stack.append(i%n)
        return res

```

```python
s = Solution()
s.nextGreaterElements([1,2,3,4,3])#[2, 3, 4, -1, 4]
```
