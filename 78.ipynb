```python
from typing import List
class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        n = len(nums)
        fin_arr = []
        def backtrack(index, temp):
            if index == n:
                fin_arr.append(temp[:])
                return 
            curr_num = nums[index]
            temp.append(curr_num)
            backtrack(index+1, temp)
            temp.pop()
            backtrack(index+1, temp)
        backtrack(0, [])
        return fin_arr
    def subsets_BFS(self, nums):
        subsets = [[]]
        for num in nums:
            new_array = []
            for subset in subsets:
                new_array.append(subset + [num])
            subsets.extend(new_array)
        print(subsets)
        
```
