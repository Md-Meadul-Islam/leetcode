```python
from typing import List
class Solution:
    def partition(self, s: str) -> List[List[str]]:
        def isPalindrome(sub):
            return sub == sub[::-1]
        final_palindrome = []
        def backtrack(s, index, curr_str, res):
            if len(s) == index:
                return final_palindrome.append(res)
            curr_str += s[index]
            if isPalindrome(curr_str):
                res.append(curr_str)
            backtrack(s, index+1, curr_str, res)
        backtrack(s, 0,"", [])

        return final_palindrome

```
