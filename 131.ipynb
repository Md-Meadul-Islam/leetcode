```python
from typing import List
class Solution:
    def partition(self, s: str) -> List[List[str]]:
        def isPalindrome(l:int, r:int):
            while l<r:
                if s[l] != s[r]:
                    return False
                l+=1
                r-=1
            return True
        n = len(s)
        final_palindrome = []
        path = []
        def backtrack(start:int):
            if start == n:
                final_palindrome.append(path[:])
                return
            for end in range(start, n):
                curr_str = s[start:end+1]
                if isPalindrome(start, end):
                    path.append(curr_str)
                    backtrack(end+1)
                    path.pop()

        backtrack(0)

        return final_palindrome

```
