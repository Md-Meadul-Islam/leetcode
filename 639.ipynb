```python
def one(c):
    if c == '0':
        return 0
    if c == '*':
        return 9
    return 1

def two(a, b):
    # "**"
    if a == '*' and b == '*':
        return 15
    
    # "*d"
    if a == '*':
        if '0' <= b <= '6':
            return 2
        if '7' <= b <= '9':
            return 1
        return 0
    
    # "d*"
    if b == '*':
        if a == '1':
            return 9
        if a == '2':
            return 6
        return 0
    
    # "dd"
    val = int(a + b)
    return 1 if 10 <= val <= 26 else 0

def numDecodings_dp(s: str) -> int:
    n = len(s)
    if n == 0 or s[0] == "0":
        return 0
    
    dp = [0]*(n+1)
    dp[0] = 1
    dp[1] = one(s[0])

    for i in range(2, n+1):
        dp[i] += one(s[i-1]) * dp[i-1]
        dp[i] += two(s[i-2], s[i-1]) * dp[i-2]
    return dp[n]  
```

```python
class Solution:
    def one(self, c):
        if c == '0':
            return 0
        if c == '*':
            return 9
        return 1

    def two(self, a, b):
        # "**"
        if a == '*' and b == '*':
            return 15
        
        # "*d"
        if a == '*':
            if '0' <= b <= '6':
                return 2
            if '7' <= b <= '9':
                return 1
            return 0
        
        # "d*"
        if b == '*':
            if a == '1':
                return 9
            if a == '2':
                return 6
            return 0
        
        # "dd"
        val = int(a + b)
        return 1 if 10 <= val <= 26 else 0

    def numDecodings(self, s: str) -> int:
        MOD = 1_000_000_007
        n = len(s)
        if n == 0 or s[0] == "0":
            return 0
        
        dp = [0]*(n+1)
        dp[0] = 1
        dp[1] = self.one(s[0])

        for i in range(2, n+1):
            dp[i] += (self.one(s[i-1]) * dp[i-1] + self.two(s[i-2], s[i-1]) * dp[i-2])% MOD
        return dp[n]
```
