```python
def numDecodings_dp(s: str) -> int:
    n = len(s)
    if n == 0 or s[0] == "0":
        return 0
    
    dp = [0]*n
    dp[0] = 1
    for i in range(1, n):
        curr_str = int(s[i])
        with_prev = int(s[i-1:i+1])
        if curr_str != 0:
            dp[i] += dp[i-1]
        if s[i] == "*":
            dp[i] += 8
        if 10 <= with_prev <=26:
            dp[i] += dp[i-2] if i>=2 else 1
    return dp[-1]  
```
