5. Longest Palindromic Substring<br>
âšªDynamic Programming

```python
class Solution:
    def longestPalindrome_brute_force(self, s: str) -> str:
        def isPalindrome(l:int, r:int):
            while l<r:
                if s[l] != s[r]:
                    return False
                l+=1
                r-=1
            return True
        n = len(s)
        max_len = 0
        max_palin = ""
        for start in range(n):
            for end in range(start, n):
                max_len = end - start + 1
                max_prev_len = len(max_palin)
                if isPalindrome(start, end) and max_len>max_prev_len:
                    max_palin = s[start:end+1]
        return max_palin
    
    def longestPalindrome_momoization(self, s: str) -> str:
        n = len(s)
        memo = {}

        def isPal(l, r):
            if l >= r:
                return True

            if (l, r) in memo:
                return memo[(l, r)]

            if s[l] == s[r] and isPal(l + 1, r - 1):
                memo[(l, r)] = True
            else:
                memo[(l, r)] = False

            return memo[(l, r)]

        start = 0
        max_len = 1

        for l in range(n):
            for r in range(l, n):
                if isPal(l, r) and (r - l + 1) > max_len:
                    start = l
                    max_len = r - l + 1

        return s[start:start + max_len]
    
    def longestPalindrome_expand_center(self, s: str) -> str:
        if len(s) <= 1:
            return s

        def expand(l, r):
            while l >= 0 and r < len(s) and s[l] == s[r]:
                l -= 1
                r += 1
            return l + 1, r - 1

        start = 0
        end = 0

        for i in range(len(s)):
            l1, r1 = expand(i, i)       # odd length
            l2, r2 = expand(i, i + 1)   # even length

            if r1 - l1 > end - start:
                start, end = l1, r1
            if r2 - l2 > end - start:
                start, end = l2, r2

        return s[start:end + 1]
    
    def longestPalindrome_expand_center_2(self, s: str) -> str:
        n = len(s)
        res = ""

        def expand(left, right):
            while left >= 0 and right < n and s[left] == s[right]:
                left -= 1
                right += 1
            return s[left + 1:right]

        for i in range(n):
            odd = expand(i, i)
            even = expand(i, i + 1)
            res = max(res, odd, even, key=len)
            
        return res

    def longestPalindrome_dp(self, s: str) -> str:
        n = len(s)
        if n <= 1:
            return s

        dp = [[False] * n for _ in range(n)]
        start = 0
        max_len = 1

        # single characters
        for i in range(n):
            dp[i][i] = True
        # length from 2 to n
        for length in range(2, n + 1):
            for l in range(0, n - length + 1):
                r = l + length - 1

                if s[l] == s[r]:
                    if length <= 2 or dp[l + 1][r - 1]:
                        dp[l][r] = True
                        if length > max_len:
                            max_len = length
                            start = l

        return s[start:start + max_len]


    
```
