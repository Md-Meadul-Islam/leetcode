```python
class Solution:
    def climbStairsRecursion(self, n: int) -> int:
        if n <= 1:
            return 1
        one = self.climbStairsRecursion(n-1)
        two = self.climbStairsRecursion(n-2)
        return one + two
    def climbStairs_Memorization_DFS(self, n:int)->int:
        memo = {}
        def dfs(n):
            if n<=1 :
                return 1
            if n in memo:
                return memo[n]
            memo[n] = dfs(n-1) + dfs(n-2)
            return memo[n]
        return dfs(n)

    def climbStairs_DP(self, n: int) -> int:
        if n <= 1:
            return 1
        prev2 = 1
        prev = 1
        for _ in range(2, n+1):
            curr = prev2 + prev
            prev2 = prev
            prev = curr
        return prev
```
