```python
class Solution:
    def numDecodings(s: str) -> int:
        n = len(s)

        def dfs(i):
            # Base case: reached end successfully
            if i == n:
                return 1
            
            # Invalid path (leading zero)
            if s[i] == '0':
                return 0
            
            # Choice 1: take one digit
            ways = dfs(i + 1)

            # Choice 2: take two digits (if valid)
            if i + 1 < n:
                two_digit = int(s[i:i+2])
                if 10 <= two_digit <= 26:
                    ways += dfs(i + 2)

            return ways

        return dfs(0)
    
    def numDecodings_dp(self, s: str) -> int:
        n = len(s)
        if n == 0:
            return 0
        if n == 1:
            return 0 if s[0] == "0" else 1
        
        dp = [0]*n
        dp[0] = 1
        for i in range(1, n):
            curr_str = int(s[i])
            with_prev = int(s[i-1:i+1])
            if curr_str != 0:
                dp[i] += dp[i-1]
            if 10 <= with_prev <=26:
                dp[i] += dp[i-2] if i>=2 else 1
        return dp[-1]        
    
    def numDecodings_tabulation(self, s:str)->int:
        n = len(s)
        if n == 0 or s[0] == "0":
            return 0
        prev2 = 1
        prev1 = 1
        for i in range(1, n):
            curr = 0
            one_digit = int(s[i])
            two_digit = int(s[i-1:i+1])
            if one_digit != 0:
                curr += prev1
            if 10 <= two_digit <=26:
                curr += prev2
            prev2= prev1
            prev1 = curr
        return prev1
```
