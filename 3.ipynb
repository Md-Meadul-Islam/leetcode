```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        hashMap = {}
        longest_substring_len = 0
        left = 0
        for i in range(len(s)):
            ch = s[i]
            if ch in hashMap and hashMap[ch]>=left:
                left = hashMap[ch] +1
            longest_substring_len = max(longest_substring_len, (i - left + 1))
            hashMap[ch] =  i
        return longest_substring_len
    
    def lengthOfLongestSubstring2(self, s:str)->int:
        hash_str = set()
        left = 0
        max_len = 0
        for i in range(len(s)):
            ch = s[i]
            print(hash_str, left)
            if ch in hash_str:
                hash_str.remove(ch)
                left += 1
            hash_str.add(ch)
            max_len = max(max_len, len(hash_str))
        return max_len


        
```
