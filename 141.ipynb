```python
class ListNode():
    def __init__(self, val):
        self.val = val
        self.next = None
```

```python
class CustomLinkList:
    def hasCycle(self, head):
        slow = head
        fast = head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            if fast == slow:
                return True
        return False
```

```python
head = ListNode(5)
head.next = ListNode(3)
head.next.next = ListNode(4)
head.next.next.next = ListNode(8)
# head.next.next.next.next = head.next
```

```python
ll = CustomLinkList()
ll.hasCycle(head)
```
