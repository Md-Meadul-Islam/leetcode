1200. Minimum Absolute Difference

```python
from typing import List

class Solution:
    def minimumAbsDifference(self, arr: List[int]) -> List[List[int]]:
        arr.sort()
        n= len(arr)
        if n < 2:
            return []
        min_diff = float('inf')
        result = []
        for i in range(1, n):
            num_1 = arr[i-1]
            num_2 = arr[i]
            diff = abs(num_1 - num_2)
            if diff < min_diff:
                min_diff = diff
                result = [[num_1, num_2]]
            elif diff == min_diff:
                result.append([num_1, num_2])
        return result
```
